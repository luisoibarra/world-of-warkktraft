# Ejercicio 2

Estas en un castillo que será asediado por una fuerza que te supera, por suerte tu salvación se encuentra a unos N días de espera. Tu misión es resistir hasta que lleguen los refuerzos. En los almacenes del castillo se tienen C cantidad de cuero, H cantidad de hierro y M cantidad de madera. Cuentas con una fuerza de A artesanos y G guerreros para defenderte.

El enemigo tardará unos D días en llegar, en ese tiempo tienes que irte preparando para la dura batalla, construye armas que puedan contener las arremetidas furiosas de los malvados que quieren tomar las vidas de tus súbditos. Los artesanos saben construir varias armas:

- Hacha: Le toma a A_1 artesano T_1 dias completarla, usando C_1 de cuero, H_1 de hierro y M_1 de madera, usada por G_1 guerreros y proporciona D_1 de daño a los enemigos
- Espada: Le toma a A_2 artesano T_2 dias completarla, usando C_2 de cuero, H_2 de hierro y M_2 de madera, usada por G_2 guerreros y proporciona D_2 de daño a los enemigos
- Catapulta: Le toma a A_3 artesano T_3 dias completarla, usando C_3 de cuero, H_3 de hierro y M_3 de madera, usada por G_3 guerreros y proporciona D_3 de daño a los enemigos

TODO: Se puede convertir guerreros a artesanos y viceversa, lo que requieren de un tiempo de adiestración de A dias.
TODO: Construir estructuras en varios dias

Los enemigos a partir del dia D comenzarán el insesante ataque.

Suerte, esperemos que no queden solo ruinas para los refuerzos.

## Definicion de variables y constantes

| Nombre | Variable/Constante/Calculable | Descripcion |
| -- | -- | -- |
| $a_t$  | Constante | Cantidad total de artesanos |
| $g_t$  | Constante | Cantidad total de guerreros |
| $aw_j$ | Constante | Artesanos necesarios para construir el arma j |
| $cw_{ij}$ | Constante | Coste del recurso i para construir el arma j |
| $ww_j$ | Constante| Cantidad de guerreros necesarios para usar el arma j |
| $E_i$ | Constante | Poder del ataque enemigo el dia i |
| $d_j$ | Constante | Daño del arma j |
| $CT_{ij}$| Calculable | Coste total en el dia i del recurso j |
| $a_{ij}$ | Variable | Artesanos asignados el dia i a construir el arma j |
| $r_{ij}$ | Variable | Recurso j en el dia i |
| $bw_{ij}$ | Variable | Armas construidas el dia i de tipo j |
| $Aw_{ij}$ | Variable | Guerreros armados el dia i con el arma j |
| $AU_{ij}$ | Variable | Armas j en uso el dia i |

## Restricciones

Para todo dia i: 

- $\sum_j a_{ij} \le a_t$ : La asignacion de artesanos no puede superar la cantidad de ellos
- $\forall_j a_{ij} = aw_j bw_{(i+1)j}$ : La asigancion de artesanos a construir el arma j tiene que satisfacer la cantidad necesaria para realizar la tarea
- $\forall_j CT_{ij} = \sum_z cw_{jz} bw_{(i+1)z} \le r_{ij}$ : Tienen que existir los recursos para construir las armas asociadas al dia
- $\forall_j r_{(i+1)j} = r_{ij} - CT_{ij} $ : Actualizacion de los recursos por dia
- $\sum_j Aw_{ij} \le g_t$ : Los guerreros asociados a las armas no pueden superar el maximo
- $\forall_j AU_{ij} ww_j = Aw_{ij} $ : Las armas tienen la cantidad de guerreros suficiente para operar
- $\sum_j AU_{ij} d_j \ge E_i $ El poder de ataque tiene que ser superior al ataque del enemigo
- $\forall_j AU_{ij} \le \sum_k^{i-1} bw_{kj}$ : Las armas tienen que existir para poder usarse

### Observaciones

Para modelar los dias de espera para el ataque se tiene que los primeros elementos de $E_i$ son ceros.

Los recursos iniciales se pueden asignar poniendo restricciones al problema tales como:

- $r_{0j} = r_j^0$

Se pueden asignar una cantidad de armas iniciales, aunque en este caso se deja en cero por defecto:

- $bw_{0,j} = bw_j^0$

Falta modelar la cantidad de dias necesarios para construir un arma, esto se puede modelar haciendo un arma fantasma y poniendo una restricción al problema que asegure la existencia de las armas fantasmas necesarias para crear el arma final. Estas armas fantasmas se modelan como armas de daño 0 y dividiendo los recursos necesarios para construirla entre los dias disponibles.

Se puede renombrar las variables de $bw_{ij}$ a $bwd_{ije}$ donde e es la etapa del proceso de construccion del arma en cuestion. La restricciones a agregar tendrian la siguiente forma:

- $ \forall_i \forall_j \sum_k^{i-1} bwd_{ij(e-1)} \ge bwd_{ije} $ : La cantidad de armas en la etapa anterior a la etapa siguiente en cada dia pasado tiene que superar la cantidad de armas en la etapa siguiente en los dias posteriores

## Funcion objetivo

Básicamente cualquier solucion factible del problema satisface el objetivo de sobrevivir, aunque se podria buscar las que minimizan o maximizan ciertos criterios. Por ejemplo:

- $\max \sum_i \sum_j AU_{ij}$ : Maximiza el poder de la defensa del castillo.
- $\min \sum_i \sum_j bw_{ij}$ : Minimiza la cantidad de armas construidas para la defensa
